# ESTUDO DE CLASSIFICADORES

```python
import pandas as pd

from sklearn.model_selection          import train_test_split
from sklearn.metrics                  import accuracy_score
from sklearn.neighbors                import KNeighborsClassifier
from sklearn.svm                      import SVC
from sklearn.preprocessing            import minmax_scale
from sklearn.tree                     import DecisionTreeClassifier
from sklearn.ensemble                 import RandomForestClassifier
```

```python
base = pd.read_csv('https://raw.githubusercontent.com/DanielAsvel/Diabete/main/diabetes.csv')
```

```python
base.shape
```

```python
x = base.drop('Outcome', axis =1)
y = base['Outcome']
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, train_size=2/3, random_state=8)
```

## KNN

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(x_treino,y_treino)

(accuracy_score(y_teste, knn.predict(x_teste)))*100
```
